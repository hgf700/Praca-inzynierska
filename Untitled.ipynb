# Early Steps of Work: Creating Test Data in csv & Testing Neural Networks

```python
import numpy as np
import random
import csv


# ====================================================
# 1. GENEROWANIE DANYCH
# ====================================================
def generate_random_schedule(employee_shifts_wanted, max_tries=1000):
    """
    PrГіbuje wygenerowaД‡ harmonogram dla zadanych 'zapotrzebowaЕ„'
    (ile zmian chce przepracowaД‡ kaЕјdy z 5 pracownikГіw).

    Zwraca:
       - 105-elementowД… listД™ (0/1) oznaczajД…cД… przypisanie pracownik->zmiana
         w ukЕ‚adzie: (dzieЕ„ 0..6) x (zmiana 0..2) x (pracownik 0..4).
       - True, jeЕ›li siД™ udaЕ‚o, False w przeciwnym razie.
    """
    num_days = 7
    num_shifts_per_day = 3
    num_emps = 5

    # Maska [dzieЕ„][zmiana] -> ktГіry pracownik (0..4) lub 5 ("nikt")
    # PoczД…tkowo wszД™dzie "nikt"
    assignment = [[5 for _ in range(num_shifts_per_day)] for _ in range(num_days)]

    # Licznik, ile zmian kaЕјdemu pracownikowi przypisaliЕ›my
    shifts_count = [0] * num_emps

    # Funkcja pomocnicza do sprawdzenia, czy moЕјna przypisaД‡ pracownika e
    def can_assign(day, shift, e):
        # 1) SprawdЕє, czy e juЕј nie pracowaЕ‚ dziЕ› w innej zmianie
        for s in range(num_shifts_per_day):
            if assignment[day][s] == e:
                return False

        # 2) SprawdЕє przerwД™ noc->rano (nocna zmiana = 2, poranna = 0)
        if shift == 0 and day > 0:
            # Czy dzieЕ„ wczeЕ›niej nocna zmiana byЕ‚a tД… samД… osobД…?
            if assignment[day - 1][2] == e:
                return False

        # 3) Czy pracownik nie przekroczy zapotrzebowania?
        if shifts_count[e] >= employee_shifts_wanted[e]:
            return False

        return True

    # SprГіbujmy "na Е›lepo" losowaД‡ przypisania
    # BД™dziemy prГіbowaД‡ rozlosowaД‡ obsadД™, dopГіki nie speЕ‚nimy warunkГіw
    # albo nie wyczerpiemy prГіb (max_tries)

    for _ in range(max_tries):
        # Wyzeruj
        for d in range(num_days):
            for s in range(num_shifts_per_day):
                assignment[d][s] = 5  # "nikt"
        shifts_count = [0] * num_emps

        # Iterujemy przez wszystkie dni i zmiany
        for d in range(num_days):
            for s in range(num_shifts_per_day):
                # Losowo permutujemy kolejnosc pracownikow + "nikt" (index 5)
                candidates = list(range(num_emps + 1))  # 0..4 + 5
                random.shuffle(candidates)

                assigned = False
                for e in candidates:
                    if e == 5:
                        # Opcja: nikt nie pracuje
                        # MoЕјemy zostawiД‡ tak i przejЕ›Д‡ dalej
                        assigned = True
                        assignment[d][s] = 5
                        break
                    else:
                        if can_assign(d, s, e):
                            # Przypisz
                            assignment[d][s] = e
                            shifts_count[e] += 1
                            assigned = True
                            break
                # JeЕ›li nie udaЕ‚o siД™ nikogo przypisaД‡ (candidates wyczerpani),
                # to wychodzimy z pД™tli i prГіbujemy od nowa
                if not assigned:
                    break
            if not assigned:
                break

        # Po skoЕ„czonej prГіbie sprawdzamy, czy wszyscy pracownicy
        # majД… DOKЕЃADNIE tyle zmian, ile chcieli
        # (jeЕ›li ma byД‡ minimalnie zamiast dokЕ‚adnie, moЕјna modyfikowaД‡ warunek)
        if shifts_count == employee_shifts_wanted:
            # UdaЕ‚o siД™
            # Zamieniamy assignment na 105 binarnych atrybutГіw
            output = []
            for d in range(num_days):
                for s in range(num_shifts_per_day):
                    for e in range(num_emps):
                        if assignment[d][s] == e:
                            output.append(1)
                        else:
                            output.append(0)
            return output, True

    # JeЕ›li nie udaЕ‚o siД™ w max_tries
    return [], False


def generate_dataset(num_samples=50, filename="harmonogram_data.csv"):
    """
    Generuje 'num_samples' przykЕ‚adГіw i zapisuje do pliku CSV.
    KaЕјdy przykЕ‚ad:
      - 5 kolumn wejЕ›ciowych (employee_shifts_wanted)
      - 105 kolumn wyjЕ›ciowych (assignment binarny)
    """
    # Przygotuj nagЕ‚Гіwek
    # Input: wanted_emp0..4
    # Output: out_0..out_104
    header = []
    for i in range(5):
        header.append(f"wanted_emp{i}")
    for i in range(105):
        header.append(f"out_{i}")

    rows = []
    generated = 0

    # Aby generowaД‡ pewne zrГіЕјnicowanie, weЕєmy w pД™tli,
    # ale moЕјna teЕј generowaД‡ do skutku, aЕј zbierzemy 50
    attempts = 0
    while generated < num_samples and attempts < 10000:
        attempts += 1

        # Losujemy, ile zmian chce kaЕјdy z 5 pracownikГіw (2-5)
        emp_wanted = [random.randint(2, 5) for _ in range(5)]

        schedule, success = generate_random_schedule(emp_wanted)
        if success:
            row = emp_wanted + schedule
            rows.append(row)
            generated += 1

    # Zapis do CSV
    with open(filename, "w", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(header)
        writer.writerows(rows)

    print(f"Wygenerowano {generated} harmonogramГіw i zapisano do {filename}.")

# ====================================================
# 2. WYWOЕЃANIE GENEROWANIA DANYCH
# ====================================================
generate_dataset(num_samples=50, filename="harmonogram_data.csv")
```

```python
# ====================================================
# 3. TRENOWANIE SIECI NEURONOWEJ
# ====================================================
# PoniЕјej przykЕ‚ad z Keras (TensorFlow).
# Musisz mieД‡ zainstalowane biblioteki: tensorflow lub keras.
# JeЕјeli nie masz вЂ“ zainstaluj:
#   pip install tensorflow
# lub
#   pip install keras
# (zaleЕјnie od wersji Pythona i Е›rodowiska).

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import pandas as pd

# Wczytanie danych z CSV
data = pd.read_csv("harmonogram_data.csv")

# PodziaЕ‚ na X (wejЕ›cie: 5 kolumn) i Y (wyjЕ›cie: 105 kolumn)
X_cols = [f"wanted_emp{i}" for i in range(5)]
Y_cols = [f"out_{i}" for i in range(105)]

X = data[X_cols].values
Y = data[Y_cols].values

# Definicja modelu
model = Sequential()
model.add(Dense(32, activation='relu', input_shape=(5,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(105, activation='sigmoid'))  # 105 wyjЕ›Д‡ binarnych
model.compile(optimizer=Adam(), loss='binary_crossentropy')

# Trening
model.fit(X, Y, epochs=50, batch_size=8, validation_split=0.2)

# Po treningu moЕјesz wyprГіbowaД‡ prognozД™:
# (oczywiЕ›cie sieД‡ moЕјe "gЕ‚upoty" przewidywaД‡ przy tak maЕ‚ej liczbie prГіbek)
test_input = np.array([[3, 4, 5, 2, 5]])
prediction = model.predict(test_input)
print("Przewidywany harmonogram (pierwsze 21 wartoЕ›ci):")
print(prediction[0][:21])
```
